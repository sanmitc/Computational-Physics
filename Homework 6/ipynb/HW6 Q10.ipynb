```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def func(vec, t):
    x=vec[0]
    y=vec[1]
    r=math.sqrt(x**2+y**2)
    return np.array([-(G*M*x)/r**3, -(G*M*y)/r**3])
```

```python
x=1.471e11
vx=0
y=0
vy=3.0287e4
t=0
h=3600
per=5*365*24*3600
G=6.674e-11
M=1.9891e30


X=[x]
Y=[y]
VxIn=[vx]
VyIn=[vy]
Vxfr=[]
Vyfr=[]
T=[t]
pos=[x,y]
vel=[vx,vy]
vel=vel + (h/2)*func(pos, t)

Vxfr.append(vel[0])
Vyfr.append(vel[1])

while(t<per):
    pos=pos + h*vel
    X.append(pos[0])
    Y.append(pos[1])
    k=h*func(pos, t)
    vel_int=vel+ k/2
    VxIn.append(vel_int[0])
    VyIn.append(vel_int[1])
    vel=vel+k
    Vxfr.append(vel[0])
    Vyfr.append(vel[1])
    t+=h
    T.append(t)
```

```python
plt.figure(figsize=(8,8))
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title('trajectory(m)')

#around 5 revolutions
```

```python
l=len(X)
m=5.9722e24
K=[]
P=[]
E=[]
for i in range(l):
    r=math.sqrt(X[i]**2+Y[i]**2)
    P.append(-G*M*m/r)
    v_sq=VxIn[i]**2+VyIn[i]**2
    K.append(0.5*m*v_sq)
    E.append(K[i]+P[i])
```

```python
plt.figure(figsize=(8,8))
plt.plot(T, K, label='KE')
plt.plot(T, P, label='PE')
plt.plot(T, E, label='Total E')
plt.xlabel("t")
plt.ylabel("Energy")
plt.title('Energy variation')
plt.legend()

#total energy constant, PE, KE vary.
```
