```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#defining the functions
w=1

def funcx(vec, t):
    x=vec[0]
    y=vec[1]
    return y

def funcy(vec, t):
    x=vec[0]
    y=vec[1]
    return -w**2*x

def func(vec,t):
    return np.array([funcx(vec,t), funcy(vec, t)])
```

```python
#defining RK for vectors
def RK_4(x, t, func, h):
    f=func(x,t)
    k1=h*f
    k2=h*func(x+k1/2, t+h/2)
    k3=h*func(x+k2/2, t+h/2)
    k4=h*func(x+k3, t+h)
    x= x + (k1+2*k2+2*k3+k4)/6
    t= t + h
    return [x,t]
```

```python
h=0.01

time=[0]
X1=[1]
X1_dash=[0]

t=0
x=1
y=0

vec=np.array([x,y])

while(t<=50):
    vec=RK_4(vec, t, func, h)[0]
    t=RK_4(vec, t, func, h)[1]
    X1.append(vec[0])
    X1_dash.append(vec[1])
    time.append(t)
```

```python
plt.figure(figsize=(8,8))
plt.plot(time, X1)
plt.xlabel("time")
plt.ylabel("x")
plt.title('SHO')
```

```python
#now making initial condition x=2

h=0.01

time=[0]
X2=[2]
X2_dash=[0]

t=0
x=2
y=0

vec=np.array([x,y])

while(t<=50):
    vec=RK_4(vec, t, func, h)[0]
    t=RK_4(vec, t, func, h)[1]
    X2.append(vec[0])
    X2_dash.append(vec[1])
    time.append(t)
```

```python
plt.figure(figsize=(8,8))
plt.plot(time, X1, label='x(0)=1')
plt.plot(time, X2, label='x(0)=2')
plt.xlabel("time")
plt.ylabel("x")
plt.title('SHO')
plt.legend()


#as seen from the graph, period remains same.
```
