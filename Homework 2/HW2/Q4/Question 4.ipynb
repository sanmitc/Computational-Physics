```python
import numpy as np
import math
import matplotlib.pyplot as plt

def bessel(m,x,N):
    val=0
    y=0
    f=0
    num=math.pi/N
    for i in range(0,N+1):
        f=math.cos(m*y-x*math.sin(y))/math.pi
        y+=num
        if(i==0):
            val+=f
        if(i==N+1):
            val+=f
        elif(i%2==0):
            val+=2*f
        elif(i%2==1):
            val+=4*f
    val=val*(num/3)
    return val

x=np.linspace(0,20,200)
j0=[]
j1=[]
j2=[]


for i in range(200):
    j0.append(bessel(0,x[i],1000))
    j1.append(bessel(1,x[i],1000))
    j2.append(bessel(2,x[i],1000))
    
plt.plot(x,j0, label='j0')
plt.plot(x,j1, label='j1')
plt.plot(x,j2, label='j2')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
    
   #part b






            
        
        
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

wavelng= 0.5 

k=((2*math.pi)/wavelng)

def rad(x,y):
    r=math.sqrt(x**2+y**2)
    kr=k*r
    return (bessel(1,kr,1000)/kr)**2

x=np.zeros((200,200))

for i in range(1,200):
    for j in range(1,200):
        if(i!=100 or j!=100):
            x[j][i]= rad((i-100)/200,(j-100)/200)
x[100][100]=0.5


plt.figure(figsize=(10,10))
plt.imshow(x, vmax=0.01)
plt.colorbar()
```
