```python
#part a

import numpy as np
import math
import matplotlib.pyplot as plt

angle=np.linspace(0,2*math.pi, 200)
x=[]
y=[]
for i in range(200):
    x.append(2*math.cos(angle[i])+math.cos(2*angle[i]))
    y.append(2*math.sin(angle[i])-math.sin(2*angle[i]))
#print(theta)

plt.plot(x,y, color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('deltoid curve')
plt.show()
```

```python
#part b

import numpy as np
import math
import matplotlib.pyplot as plt

angle=np.linspace(0,10*math.pi, 500)

r=[]
x=[]
y=[]

for i in range(500):
    r.append(angle[i]**2)
    x.append(r[i]*math.cos(angle[i]))
    y.append(r[i]*math.sin(angle[i]))
    
plt.plot(x,y, color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gallilean Spiral')
plt.show()
```

```python
#part c

import numpy as np
import math
import matplotlib.pyplot as plt

angle=np.linspace(0,24*math.pi, 2000)

r=[]
x=[]
y=[]

for i in range(2000):
    r.append(np.exp(math.cos(angle[i])) - 2*math.cos(4*angle[i]) + (math.sin(angle[i]/12))**5)
    x.append(r[i]*math.cos(angle[i]))
    y.append(r[i]*math.sin(angle[i]))
    
plt.plot(x,y, color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Fey's function")
plt.show()
```
