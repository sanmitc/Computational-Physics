```python
#question 16
import numpy as np
import math
import matplotlib.pyplot as plt



r=np.linspace(1,4,300)
y=[1/2]*300
for k in range(2000): 
    for j in range(300):
        y[j]=r[j]*y[j]*(1-y[j])
    if(k>1000):
        plt.scatter(r,y,0.1)
plt.xlabel('r')
plt.ylabel('x')
plt.title('deterministic chaos')
plt.show()
```
