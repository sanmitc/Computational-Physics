```python
#question 18
import numpy as np
import math
import matplotlib.pyplot as plt



r=np.linspace(1,4,300)
y=[1/2]*300
for k in range(2000): 
    for j in range(300):
        y[j]=r[j]*y[j]*(1-y[j])
    if(k>1000):
        plt.scatter(r,y,0.1)
plt.xlabel('r')
plt.ylabel('x')
plt.title('deterministic chaos')
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

file=open(r"millikan.txt")

x=[]
y=[]

for rows in file:
    rows=rows.split(' ')
    x.append(float(rows[0]))
    y.append(float(rows[1]))
    
plt.scatter(x,y)
   
E_x=0
E_y=0
E_xx=0
E_xy=0

n=len(x)

for i in range(n):
    E_x+=x[i]
    E_xx+=x[i]**2
    E_y+=y[i]
    E_xy+=x[i]*y[i]
    
E_x=E_x/n
E_y=E_y/n
E_xx=E_xx/n
E_xy=E_xy/n
    
m=(E_xy-E_x*E_y)/(E_xx-E_x**2)
c=(E_xx*E_y-E_x*E_xy)/(E_xx-E_x**2)

#print(m,c)
    
y_new=[]

for i in range(n):
    y_new.append(m*x[i]+c)
    
plt.plot(x,y_new,'-')

plt.xlabel("x")
plt.ylabel("y")
plt.title('Millikan Data')
    
    
```
